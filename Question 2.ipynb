```python

import requests
from bs4 import BeautifulSoup
import pandas as pd
import time


search_set = ['big+tech', 'google', 'amazon', 'stocks', 'covid', 
              'vaccine', 'pfizer', 'trump', 'biden', 'senate']

art_hyperlink_list = []
readtime_list = []
article_title_list = []
dateandtime_list = []

for search in search_set:
    URL = 'https://www.reuters.com/search/news?blob=' + search 
    #print('Now Scraping:', URL)
    time.sleep(3)
    page = requests.get(URL)
    soup = BeautifulSoup(page.content, 'html.parser')
    tags = soup.find_all('div', class_= 'search-result-content')
    
    for tag in tags:
        
        try:
            article_title = tag.find('h3', class_= 'search-result-title').get_text().strip()
        except:
            article_title = 'NA'
        #print(article_title)
        article_title_list.append(article_title)
        
        try:
            dateandtime = tag.find('h5', class_= 'search-result-timestamp').get_text().strip()
        except:
            dateandtime = 'NA'
        #print(dateandtime)
        dateandtime_list.append(dateandtime)
        
        try:
            art_hyperlink = tag.find('a').get('href')
            fullURL = 'https://www.reuters.com' + art_hyperlink
        except:
            art_hyperlink = 'NA'
        #print(art_hyperlink)
        art_hyperlink_list.append(art_hyperlink)
        
        articlepage = requests.get(fullURL)
        soup2 = BeautifulSoup(articlepage.content, 'html.parser')
        tags2 = soup2.find_all('div', class_= 'ArticleBodyWrapper')
        
        for t in tags2:
            
            try:
                readtime = t.find('p', class_= 'TextLabel__text-label___3oCVw TextLabel__gray___1V4fk TextLabel__small-all-caps-spaced-out___3O9H4 ReadTime-read-time-1s3CG ArticleBody-read-time-29pGN').get_text().strip(' Read')
            except:
                readtime = 'NA'
            #print(readtime)
            readtime_list.append(readtime)
                 
data_dict = {'Headline':article_title_list, 
            'Dateandtime':dateandtime_list,
            'Hyperlink':art_hyperlink_list,
            'Readtime':readtime_list}

arts = pd.DataFrame(data_dict)

print(arts)

arts.to_csv('question_2.csv', index = True, header = True)          
```
